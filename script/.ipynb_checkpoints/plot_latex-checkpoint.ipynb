```python
### import libraries ###
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from singlestarevo import extract     # from singlestarevo.py, to extract single stellar evolution data
```

```python
### set parameters ###

# [Nsim] is the number of simulations and is used to locate all the related folders
# [Z] is the metallicity of the stars                               e.g. '02' for Z=0.02 solar metallicity
# [SN] is the SN explosion prescription adopted                     e.g. 'del' for delayed or 'com' for compactness
# [thres_RL] is the threshold for Roche Lobe filling binaries
# [thres_wind] is the threshold for wind fed binaries

Nsim = '1mln'
Z_list = ['02', '002', '02', '002']
SN_list = ['del','del','com','com']
thres_RL = 1.
thres_wind = 0.8

path_plots=f'./'          # where plots will be stored
```

```python
### set parameters for plots ###
plt.rcParams.update({'text.usetex': True, 
                     'font.family': 'Helvetica', 
                     'font.size': 40})
lw = 2.5  # linewidth
```

```python
plottype = 'hist_mass_prog_rem'

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(30,20))

for Z,SN,ax in zip(Z_list,SN_list,axs.flat):
    ### set paths ###
    path = f'{Nsim}_Z{Z}_{SN}'             # folder path identifier for a specific simulation run
    path_df=f'./{path}/dataframes/'        # where processed dataframes are stored
    
    # extract data
    p1,r1 = pd.read_csv(f'{path_df}p_BHBH_bound_WRBH.csv'), pd.read_csv(f'{path_df}r_BHBH_bound_WRBH.csv')
    p2,r2 = pd.read_csv(f'{path_df}p_BHBH_GW_WRBH.csv'), pd.read_csv(f'{path_df}r_BHBH_GW_WRBH.csv')

    prog_list = [p1,p2]  # progenitor list
    rem_list = [r1,r2]   # remnant list


    labels1=['bBBH prog',
            r'GW-BBH prog']
    labels2=['bBBH rem',
            r'GW-BBH rem']
    colors1=['k','steelblue']
    colors2=['lightgrey','cadetblue']
    linestyles=['dashed','solid']
    
    bin_edges_mass=np.arange(0,151,1)       # 1 M_sun is the witdth of each bin

    for prog,rem, label1,label2, color1,color2,linestyle in zip(prog_list, rem_list, labels1,labels2, colors1,colors2,linestyles):

        ax.hist(rem['Mass_0'], bins=bin_edges_mass, label=label2, color=color2, linestyle=linestyle, linewidth = lw, log=True,  histtype='stepfilled')
        ax.hist(prog['Mass_0'], bins=bin_edges_mass, label=label1, color=color1, linestyle=linestyle, linewidth = lw, log=True,  histtype='step')
    
    ax.annotate(f'Z=0.{Z} \n SN={SN}', xy=(0.8, 0.8),xycoords='axes fraction')
    ax.set_xlabel(r'$\displaystyle M_1~[M_\odot]$')
    ax.set_ylabel(r'Number of systems')
    ax.set_ylim(ymax=20000)
    ax.set_xticks(np.arange(0, 150+1,20))

handles, labels = axs[0][0].get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol=4, bbox_to_anchor=(0.48,0.02))   
plt.tight_layout()
fig.subplots_adjust(bottom=0.15)

plt.show()
fig.savefig(f'{path_plots}def_{plottype}.png')
```

```python
plottype = 'hist_per_prog_rem'

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(30,20))

for Z,SN,ax in zip(Z_list,SN_list,axs.flat):
    ### set paths ###
    path = f'{Nsim}_Z{Z}_{SN}'             # folder path identifier for a specific simulation run
    path_df=f'./{path}/dataframes/'        # where processed dataframes are stored
    
    # extract data
    p1,r1 = pd.read_csv(f'{path_df}p_BHBH_bound_WRBH.csv'), pd.read_csv(f'{path_df}r_BHBH_bound_WRBH.csv')
    p2,r2 = pd.read_csv(f'{path_df}p_BHBH_GW_WRBH.csv'), pd.read_csv(f'{path_df}r_BHBH_GW_WRBH.csv')

    prog_list = [p1,p2]  # progenitor list
    rem_list = [r1,r2]   # remnant list


    labels1=['bBBH prog',
            r'GW-BBH prog']
    labels2=['bBBH rem',
            r'GW-BBH rem']
    colors1=['k','midnightblue']
    colors2=['lightgrey','steelblue']
    linestyles=['dashed','solid']
    
    bin_edges_period=np.logspace(np.log10(0.00001),np.log10(10000),100)  # 100 yrs width of each bin


    for prog,rem, label1,label2, color1,color2,linestyle in zip(prog_list, rem_list, labels1,labels2, colors1,colors2,linestyles):

        ax.hist(rem['Period'], bins=bin_edges_period, label=label2, color=color2, linestyle=linestyle, linewidth = lw, log=True,  histtype='stepfilled')
        ax.hist(prog['Period'], bins=bin_edges_period, label=label1, color=color1, linestyle=linestyle, linewidth = lw, log=True,  histtype='step')
    ax.axvline(x=1./365, linestyle='dotted', linewidth =lw, color='k') # 1 day in yrs unit of measure
    ax.axvline(x=1./(365*24), linestyle='dashdot', linewidth =lw, color='k') # 1 hour in yrs unit of measure
    
    ax.annotate(f'Z=0.{Z} \n SN={SN}', xy=(0.8, 0.8),xycoords='axes fraction')
    ax.set_xlabel(r'$\displaystyle \mathcal{P}$ [yrs]')
    ax.set_ylabel(r'Number of systems')
    ax.set_ylim(ymax=5000)    
    ax.set_xscale('log')

handles, labels = axs[0][0].get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol=4, bbox_to_anchor=(0.48,0.02))   
plt.tight_layout()
fig.subplots_adjust(bottom=0.15)

plt.show()
fig.savefig(f'{path_plots}def_{plottype}.png')
```

```python
plottype = 'scatter2D_prog_rem1'

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(30,20))

for Z,SN,ax in zip(Z_list,SN_list,axs.flat):
    ### set paths ###
    path = f'{Nsim}_Z{Z}_{SN}'             # folder path identifier for a specific simulation run
    path_df=f'./{path}/dataframes/'        # where processed dataframes are stored

    p1,r1 = pd.read_csv(f'{path_df}p_BHBH_bound_WRBH.csv'), pd.read_csv(f'{path_df}r_BHBH_bound_WRBH.csv')
    p2,r2 = pd.read_csv(f'{path_df}p_BHBH_GW_WRBH.csv'), pd.read_csv(f'{path_df}r_BHBH_GW_WRBH.csv')
    singlestar = extract(SN,Z)

    prog_list = [p1,p2]  # progenitor list
    rem_list = [r1,r2]   # remnant list


    edges = [None,'k']
    colors = ['grey','lime']
    markers = ['*','o']
    sizes = [2,100]
    labels = ['bBBH','GW-BBH']
    
    for prog,rem,marker,edge,color,size,label in zip(prog_list,rem_list,markers,edges,colors,sizes,labels):
        prog=prog.sort_values(by='ID')
        rem=rem.sort_values(by='ID')
        ax.scatter(prog['Mass_0'],rem['Mass_0'] , marker=marker, edgecolor = edge, c=color,s=size, label=label)

    ax.plot(singlestar['Mass'],singlestar['Mrem'], linestyle='solid', linewidth =lw, color='r', 
            alpha=0.5, label='Single stellar evolution')

        
    ax.annotate(f'Z=0.{Z} \n SN={SN}', xy=(0.08, 0.8),xycoords='axes fraction')
    ax.set_xlabel(r'$\displaystyle M_{1,\rm prog}~[M_\odot]$')
    ax.set_ylabel(r'$\displaystyle M_{1,\rm rem}~[M_\odot]$')  
    ax.set_xticks(np.arange(0, 150+1,20))
    ax.set_yticks(np.arange(0, 50+1,10))
    ax.grid(alpha=0.3)



handles, labels = axs[0][0].get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol=4, bbox_to_anchor=(0.48,0.02))   
plt.tight_layout()
fig.subplots_adjust(bottom=0.15)

plt.show()
fig.savefig(f'{path_plots}def_{plottype}.png')
```

```python
plottype = 'CE_mass'

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(30,20))
cmap= cm.get_cmap('inferno_r',2)

for Z,SN,ax in zip(Z_list,SN_list,axs.flat):
    ### set paths ###
    path = f'{Nsim}_Z{Z}_{SN}'             # folder path identifier for a specific simulation run
    path_df=f'./{path}/dataframes/'        # where processed dataframes are stored
    
    # extract data
    p1,r1 = pd.read_csv(f'{path_df}p_BHBH_GW_WRBH.csv'), pd.read_csv(f'{path_df}r_BHBH_GW_WRBH.csv')
    p2,r2 = pd.read_csv(f'{path_df}p_BHBH_GW_WRBH_wind.csv'), pd.read_csv(f'{path_df}r_BHBH_GW_WRBH_wind.csv')
    p3,r3 = pd.read_csv(f'{path_df}p_BHBH_GW_WRBH_RL.csv'), pd.read_csv(f'{path_df}r_BHBH_GW_WRBH_RL.csv')
    prog_list = [p1,p2,p3]  # progenitor list
    rem_list = [r1,r2,r3]   # remnant list

    markers = ['o','v','^']
    sizes = [80,500,500]
    labels = [fr'$ R/RL < {thres_wind} $', fr'$ R/RL \geq {thres_wind} $',fr'$R/RL \geq {thres_RL} $']
    
    for prog,rem, label, marker, size in zip(prog_list, rem_list, labels,markers,sizes):
        prog, rem = prog.sort_values(by='ID'), rem.sort_values(by='ID')
        
        cs = ax.scatter(rem['Mass_0'], rem['Mass_1'], marker=marker, edgecolors= 'k',c=prog['NCE'],cmap=cmap, vmin=1, vmax=2,s=size, label=label)
        ax.plot([3,35],[3,35], color='r', alpha=0.3, linewidth=4)

    ax.annotate(f'Z=0.{Z} \n SN={SN}', xy=(0.08, 0.8),xycoords='axes fraction')
    ax.set_xlabel(r'$\displaystyle M_{1,\rm rem} [M_\odot]$')
    ax.set_ylabel(r'$\displaystyle M_{2,\rm rem} [M_\odot]$') 
    ax.set_xlim(0,40)
    ax.set_ylim(0,40)
    ax.grid(alpha=0.3)

cbar=plt.colorbar(cs, ax=axs[:,1],orientation='vertical')
cbar.set_label(r'$N_{CE}$')
cbar.set_ticks([0,1,2])

handles, labels = axs[0][0].get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol=4, bbox_to_anchor=(0.48,0.02))   
#plt.tight_layout()
fig.subplots_adjust(bottom=0.15, top=0.93, left=0.1, right=0.8)

plt.show()
fig.savefig(f'{path_plots}def_{plottype}.png')
```
