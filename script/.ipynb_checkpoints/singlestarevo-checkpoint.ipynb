```python
### import libraries ###
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
### function to extract values from aa SN model with a Z metallicity ###
# [Z] is the metallicity of the stars                           e.g. '02' for Z=0.02 solar metallicity
# [SN] is the SN explosion prescription adopted                 e.g. 'del' for delayed or 'com' for compact

def extract(SN,Z):
    ### set paths ###
    path_to_sevn_output = f'./singlestar/singlestar_Z{Z}_{SN}/sevn_output/'
    path_evolved = path_to_sevn_output + 'evolved_0.dat'  # evolved binaries
    path_log = path_to_sevn_output + 'logfile_0.dat'      # logfile
    path_out = path_to_sevn_output + 'output_0.csv'       # output file
    
    ### extract SN events from the logfile ###
    # logfile SN lines have the following structure
    # S;name;ID;SN;time;M_tot:M_He:M_CO:M_rem:rem_type:other_infos_on_natal_kicks
    # where M_tot,M_He,M_CO,M_rem,rem_type are before the SN explosion
    regex_str=r'S;\d+;(\d+);SN;\d+.\d+;\d+.\d+:(\d+.\d+):(\d+.\d+):(\d+.\d+):(\d+):'
    with open(path_log,"r") as f:
           SN_mask = re.findall(regex_str,f.read())             #Find al the SN occurence and save the ID
    col_names = ['ID','MHe','MCO','Mrem','rem_type']
    SN = pd.DataFrame(SN_mask, columns=col_names,dtype=float)   # convert into pandas dataframe

    ## read evolved and output files ##
    evolved = pd.read_csv(path_evolved, sep='\s+').rename(columns = {'#ID':'ID'}) # contains 'Mass' ZAMS
    
    # progenitors and remnants f stars that explode as SN
    merged = pd.merge(evolved,SN,on='ID')
    merged = merged[merged['rem_type'] >= 4]           # take only NS and BH remnants
    merged = merged.sort_values(by='Mass')    # sort in order to allow plotting and not scattering
    
    return merged
```

```python
### extract dataframes of interest ###
del_02 = extract('del','02')
del_002 = extract('del','002')
com_02 = extract('com','02')
com_002 = extract('com','002')
```

```python
### set parameters for plots ###
plt.rcParams.update({'text.usetex': True, 
                     'font.family': 'Helvetica', 
                     'font.size': 40})
lw = 2.5 # linewidth of histograms
```

```python
plottype = 'singleevoZ02'

bin_edges=np.arange(0,21,0.5)       # Z02

fig= plt.figure(figsize=(15,10))
plt.title(r'$Z$ = 0.02')
plt.hist(com_02['Mrem'],bins=bin_edges, density=True,log=True,
              histtype='stepfilled', color='lightgrey', linestyle='dotted', linewidth=lw, edgecolor='k',label=r'Compact')
plt.hist(del_02['Mrem'],bins=bin_edges, density=True,log=True,
               histtype='step', color='k', linestyle='solid', linewidth=lw, label=r'Delayed')

plt.xlabel(r'$\displaystyle M_{\rm rem}~[ M_\odot]$')
plt.ylabel('PDF')
plt.legend()

plt.tight_layout()
#plt.show()

fig.savefig(f'{plottype}.png')
```

```python
plottype = 'singleevo'

lw=5  # thicker linewidth for this plot
fig= plt.figure(figsize=(15,10))
plt.plot(del_02['Mass'],  del_02['MHe'], linewidth =lw, linestyle = 'dashed', c='r',label='He, Z=0.02')
plt.plot(del_02['Mass'],  del_02['MCO'], linewidth =lw, linestyle = 'solid', c='dodgerblue',label='CO, Z=0.02')

plt.plot(del_002['Mass'],  del_002['MHe'], linewidth =lw, linestyle = 'dotted', c='r',label='He, Z=0.002')
plt.plot(del_002['Mass'],  del_002['MCO'], linewidth =lw, linestyle = 'dashdot', c='dodgerblue',label='CO, Z=0.002')

plt.axhline(y=11, linestyle = 'dotted', c='k')

plt.xlabel(r'$\displaystyle M_{\rm ZAMS}~[ M_\odot]$')
plt.ylabel(r'$\displaystyle M_\odot$')
plt.legend()

plt.tight_layout()
#plt.show()

fig.savefig(f'{plottype}.png')
```
