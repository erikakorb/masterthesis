# Preliminary

```python
### import libraries ###
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
import matplotlib.lines as mlines
import os
from singlestarevo import extract     # from singlestarevo.py, to extract single stellar evolution data
```

```python
### set parameters ###

# [Nsim] is the number of simulations and is used to locate all the related folders
# [Z] is the metallicity of the stars                               e.g. '02' for Z=0.02 solar metallicity
# [SN] is the SN explosion prescription adopted                     e.g. 'del' for delayed or 'com' for compact
# [Hsup] is the threshold for spectroscopic WR                      e.g. '0.3' for WR with H < 0.3 on surface
# [thres_RL] is the threshold for Roche Lobe filling binaries
# [thres_wind] is the threshold for wind fed binaries
# [observed] to select binaries similar to one observed             e.g. 'cyg_x-3--Zd13' for Zdziarski 2013 ranges

Nsim = '1mln'
Z_list = ['02', '02', '02', '02']
SN_list = ['del','del','com','com']
Hsup = 0.3
thres_RL = 1.
thres_wind = 0.8
candidate = 'cyg_x-3'
version = '3.0.0-Spindevel'                            # for SEVN2 version adopted


wind_perc = int(thres_wind*100) # for plot labels
```

```python
### Choose if plot during all the WRBH phase or only during the observable configuration ###
# the duration parameter could be while or all
# [duration] = 'observable' means only while in the observable configuration
# [duration] = 'WRBH' means only in the whole WRBH configuration, of which the observable is only a little portion
duration = 'observable'
```

```python
### Set paths ###
path_plots=f'./v_{version}/plots/'                             # where plots will be stored
path_plots_observed = f'{path_plots}{candidate}_compare/while_{duration}/'        # plots with only observable candidates binaries

### check directories exist or create them ###
os.makedirs(path_plots_observed, exist_ok=True)
```

```python
### Prepare lists for plots ###
prefix_list = ['pr','if','pr','if']
mranges = ['Ko17','An22','Zd13']
markers_ranges = ['s','^','o']
sizes_ranges = [500,400,300]
colors_ranges = ['grey','fuchsia','lime']
```

# Prepare plots

```python
### set parameters for plots ###
plt.rcParams.update({'text.usetex': True, 
                     'font.family': 'Helvetica', 
                     'font.size': 40})
lw = 2.5  # linewidth

if candidate == 'cyg_x-3--Zd13':
    obs = 'Cyg X-3 (Zd13)'
elif candidate == 'cyg_x-3--An22':
    obs = 'Cyg X-3 (An22)'
elif candidate == 'cyg_x-3--Ko17':
    obs = 'Cyg X-3 (Ko17)'
elif candidate == 'cyg_x-3':
    obs = 'Cyg X-3'
```

# Plots

```python
plottype = 'scatter_2D_mass_WRBH_obs'

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(30,20))
fig.suptitle(f'Mass of remnant vs mass of the progenitor for GW-BBH from WRBH {obs} candidates')

for mrange, marker_range,size_range,color_range in zip(mranges,markers_ranges,sizes_ranges,colors_ranges):
    observed = f'{candidate}--{mrange}'
    
    for Z,SN,ax in zip(Z_list,SN_list,axs.flat):
        ### set paths ###
        path = f'./v_{version}/{Nsim}_Z{Z}_{SN}'             # folder path identifier for a specific simulation run
        path_df=f'./{path}/dataframes/'        # where processed dataframes are stored

        path_df_prog = f'./{path}/dataframes/progenitors/'        # where progenitor dataframes are stored
        path_df_rem = f'./{path}/dataframes/remnants/'            # where remnant dataframes are stored

        prog,rem = pd.read_csv(f'{path_df_prog}p_BHBH_GW_WRBH_{observed}.csv'), pd.read_csv(f'{path_df_rem}r_BHBH_GW_WRBH_{observed}.csv')

        singlestar = extract(SN,Z)

        prog=prog.sort_values(by='ID')
        rem=rem.sort_values(by='ID')

        if (ax == axs[0][0]) | (ax == axs[1][0]):
            ax.scatter(prog[f'Mass_0'],rem[f'Mass_0'] , marker=marker_range, edgecolor = 'k', c=color_range,s=size_range, label=mrange)
            ax.set_xlabel(r'$\displaystyle M_{1,\rm prog}~[M_\odot]$')
            ax.set_ylabel(r'$\displaystyle M_{1,\rm rem}~[M_\odot]$') 

        elif (ax == axs[0][1]) | (ax == axs[1][1]):
            ax.scatter(prog[f'Mass_1'],rem[f'Mass_1'] , marker=marker_range, edgecolor = 'k', c=color_range,s=size_range, label=mrange)
            ax.set_xlabel(r'$\displaystyle M_{2,\rm prog}~[M_\odot]$')
            ax.set_ylabel(r'$\displaystyle M_{2,\rm rem}~[M_\odot]$') 
            
for ax in axs.flat:
    ax.plot(singlestar['Mass'],singlestar['Mrem'], linestyle='solid', linewidth =lw, color='r', 
            alpha=0.2, label='Single stellar evolution')

    ax.annotate(f'Z=0.{Z} \n SN={SN}', xy=(0.08, 0.8),xycoords='axes fraction')
    ax.set_xticks(np.arange(0, 50+1,10))
    ax.set_yticks(np.arange(0, 17+1,2))
    ax.set_xlim(0,50)
    ax.set_ylim(0,17)
    ax.grid(alpha=0.3)


axs[0][0].set_title(r'$M_1$')
axs[0][1].set_title(r'$M_2$')

handles, labels = axs[0][0].get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol=4, bbox_to_anchor=(0.48,0.02))   
plt.tight_layout()
fig.subplots_adjust(wspace = 0.3, bottom=0.2)

plt.show()
fig.savefig(f'{path_plots_observed}{plottype}.jpg')
```

```python
plottype = 'per_WRBH_obs'

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(30,20))
fig.suptitle(f'Initial and final period of WRBH {obs} candidates')

for mrange, marker_range,size_range,color_range in zip(mranges,markers_ranges,sizes_ranges,colors_ranges):
    observed = f'{candidate}--{mrange}'
    
    for Z,SN,prefix,ax in zip(Z_list,SN_list,prefix_list,axs.flat):
        ### set paths ###
        path = f'./v_{version}/{Nsim}_Z{Z}_{SN}'             # folder path identifier for a specific simulation run
        path_df=f'./{path}/dataframes/'        # where processed dataframes are stored


        if (ax == axs[0][0]) | (ax == axs[1][0]):
            path_df_prog = f'./{path}/dataframes/progenitors/'        # where progenitor dataframes are stored
            path_df_rem = f'./{path}/dataframes/remnants/'            # where remnant dataframes are stored

            prog,rem = pd.read_csv(f'{path_df_prog}p_BHBH_GW_WRBH_{observed}.csv'), pd.read_csv(f'{path_df_rem}r_BHBH_GW_WRBH_{observed}.csv')
            prog=prog.sort_values(by='ID')
            rem=rem.sort_values(by='ID')
            ax.scatter(prog['Period'],rem['Period']*24*365 ,  marker=marker_range, edgecolor = 'k', c=color_range,s=size_range, label=mrange)
            ax.set_xlabel(r'$\displaystyle  \log P_{\rm prog}$ [yrs]')
            ax.set_ylabel(r'$\displaystyle  P_{\rm rem}$ [hrs]')
            ax.set_xscale('log')
            ax.set_xlim(xmin=1e-4, xmax=1e2) 
            ax.set_ylim(ymin=0, ymax=50)  

        elif (ax == axs[0][1]) | (ax == axs[1][1]):
            path_df_in = f'./{path}/dataframes/initial_WRBH/'        # where initial timesteps of WRBH are stored
            path_df_fin = f'./{path}/dataframes/final_WRBH/'         # where final timesteps of WRBH are stored

            i,f = pd.read_csv(f'{path_df_in}i_BHBH_GW_WRBH_{observed}.csv'), pd.read_csv(f'{path_df_fin}f_BHBH_GW_WRBH_{observed}.csv')
            i=i.sort_values(by='ID')
            f=f.sort_values(by='ID')
            ax.scatter(i['Period']*24*365,f['Period']*24*365 ,  marker=marker_range, edgecolor = 'k', c=color_range,s=size_range, label=mrange)
            ax.set_xlabel(r'$\displaystyle  P_{\rm in}$ [hrs]')
            ax.set_ylabel(r'$\displaystyle  P_{\rm fin}$ [hrs]')
            ax.set_xlim(xmin=4.4, xmax=5.2) 
            ax.set_ylim(ymin=4.4, ymax=5.2)  
            ax.axvspan(4.5, 5.1,0,1,facecolor='dodgerblue',alpha=0.05)
            
        ax.annotate(f'Z=0.{Z} \n SN={SN}', xy=(0.05, 0.1),xycoords='axes fraction')
        

for ax in axs.flat:
    ax.axhline(y=24, linestyle='dotted', linewidth =lw, color='k', label='1 day') # 1 day 
    ax.axhline(y=1., linestyle='dashdot', linewidth =lw, color='k', label='1 hour') # 1 hour 

    ax.axvline(x=1./(365), linestyle='dotted', linewidth =lw, color='k') # 1 day 
    ax.axvline(x=1./(24*365), linestyle='dashdot', linewidth =lw, color='k') # 1 hour

    ax.axhspan(4.5, 5.1, 0 ,1, facecolor='dodgerblue',alpha=0.3)
    ax.grid(alpha=0.3)

axs[0][0].set_title('Progenitors and remnants')
handles, labels = axs[0][0].get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol=5, bbox_to_anchor=(0.48,0.02))   # 0.48

axs[0][1].set_title('Begin and end of {observable} phase')

plt.tight_layout()
fig.subplots_adjust(wspace = 0.3,bottom=0.2)

plt.show()
fig.savefig(f'{path_plots_observed}{plottype}.jpg')
```
