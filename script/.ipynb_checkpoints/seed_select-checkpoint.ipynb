```python
import numpy as np
import pandas as pd
import os
```

```python
# Dataframe with binaries of interest
path = './1mln_Z02_com/'
df_name = 'BHBH_GW_WRBH_cyg_x-3'
path_to_df = f'{path}dataframes/progenitors/p_{df_name}.csv'
df = pd.read_csv(path_to_df)
```

```python
# create new folder to store results
os.makedirs(path+df_name, exist_ok=True)
```

```python
df
```

```python
# Select only columns of interest
cols = df.columns.to_list()
filtered = df[cols[4:-2]]
```

```python
# Filter and write the listBin2.dat file with the SEN2 input syntax
path_to_listBin2 = f'{path}{df_name}/listBin2.dat'
if os.path.exists(path_to_listBin2):
        os.remove(path_to_listBin2)
filtered.to_csv(path_to_listBin2,sep='\t',index=False,header=False)
```

```python
###### Select from evolved file a complete run ###
path_to_evolved = 'evolved_0.dat'
df = pd.read_csv(path_to_evolved, sep='\s+')
cols = df.columns.to_list()
filtered = df[cols[2:]]
filtered.to_csv('listBinSeed.dat',sep='\t',index=False,header=False)
```
