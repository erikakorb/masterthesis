```python
import numpy as np
import matplotlib.pyplot as plt

Rsun = float(7e8)  #m   700 000 km
Msun = float(2e30) #kg
G = 6.67e-11 #m**3 Kg**-1 s**-2
Gsun = Msun * G/(Rsun**3 )  # Rsun**3 Msun**-1 s**-2


M1 = np.arange(11,101,1)
Mtot = np.zeros(90) + 130
M2 = Mtot - M1

L = 1.4
a = (Mtot/Gsun) * (L/(M1*M2))**2
q = M1/M2





# set parameters
plt.rcParams.update({'text.usetex': True, 
                     'font.family': 'Helvetica', 
                     'font.size': 25})

scale= 3
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(1,1,1)
ax.plot(q,a, color='k',linewidth=4,label=r'$a$')
ax.plot(q,M1*scale, color='dodgerblue', linestyle='dashdot',linewidth=4,label=r'$M_1$')
ax.plot(q,M2*scale, color='red', linestyle='dashed',linewidth=4,label=r'$M_2$')

secaxy=ax.secondary_yaxis('right',functions=(lambda x: x/scale, lambda x: x*scale))
secaxy.set_ylabel(r'Mass [$M_\odot$]')
secaxy.ticklabel_format(axis="y")

ax.legend(loc='upper right',ncol=3)
ax.set_xlabel(r'$q=M_1/M_2$')
ax.set_ylabel(r'$a~ [R_\odot]$')

plt.show()
fig.savefig('../latex/images/masstransfer.pdf')
```
