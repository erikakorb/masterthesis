```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
# set parameters
plt.rcParams.update({'text.usetex': True, 
                     'font.family': 'Helvetica', 
                     'font.size': 25})
```

```python
# set paths
path_tables = '/home/erika/Scrivania/sevn/SEVN2-3.0.0-Spindevel_RLO/tables'
path_AGB = f'{path_tables}/SEVNtracks_parsec_AGB'
path_pureHE = f'{path_tables}/SEVNtracks_parsec_pureHe36'
```

```python
lines = np.arange(25,29,1)
#lines = [65] 12 MSUN
with open(f'{path_AGB}/0014/mass.dat', 'r') as fin:
    reader=csv.reader(fin)
    mass=np.array([[float(s) for s in row[0].split()] for i,row in enumerate(reader) if i in lines])
with open(f'{path_AGB}/0014/mhe.dat', 'r') as fin:
    reader=csv.reader(fin)
    mhe=np.array([[float(s) for s in row[0].split()] for i,row in enumerate(reader) if i in lines])
with open(f'{path_AGB}/0014/hsup.dat', 'r') as fin:
    reader=csv.reader(fin)
    hsup=np.array([[float(s) for s in row[0].split()] for i,row in enumerate(reader) if i in lines])
with open(f'{path_AGB}/0014/radius.dat', 'r') as fin:
    reader=csv.reader(fin)
    radius=np.array([[float(s) for s in row[0].split()] for i,row in enumerate(reader) if i in lines])
with open(f'{path_AGB}/0014/time.dat', 'r') as fin:
    reader=csv.reader(fin)
    time=np.array([[float(s) for s in row[0].split()] for i,row in enumerate(reader) if i in lines])

for i,line in enumerate(lines):
    fig, axs = plt.subplots(nrows=3,ncols=1, sharex=True,figsize=(8,20))

    axs[0].set_title(f'M ZAMS = {mass[i][0]}, Z=0.014')
    axs[0].plot(time[i],radius[i])
    axs[0].set_yscale('log')
    axs[0].grid(linestyle='dotted',alpha=0.8)
    #axs[0].set_xlabel('Time from ZAMS [Myr]')
    axs[0].set_ylabel('Radius [Rsun]')
    #plt.ylim(0,0.8)
    #plt.show()

    axs[1].plot(time[i],mass[i])
    axs[1].plot(time[i],mhe[i])
    axs[1].grid(linestyle='dotted',alpha=0.8)
    #axs[1].set_xlabel('Time from ZAMS [Myr]')
    axs[1].set_ylabel('Mass [Msun]')
    
    
    axs[2].plot(time[i],hsup[i])
    axs[2].grid(linestyle='dotted',alpha=0.8)
    axs[2].set_xlabel('Time from ZAMS [Myr]')
    axs[2].set_ylabel('Hsup')
    #plt.ylim(0,0.8)
    plt.show()
    fig.savefig(f'./parsectracks/AGB_Z014_{int(mass[i][0])}.png')
```

```python
lines = np.arange(55,75,1)
#lines = [65] 12 MSUN
with open(f'{path_pureHE}/001/mass.dat', 'r') as fin:
    reader=csv.reader(fin)
    mass=np.array([[float(s) for s in row[0].split()] for i,row in enumerate(reader) if i in lines])
with open(f'{path_pureHE}/001/radius.dat', 'r') as fin:
    reader=csv.reader(fin)
    radius=np.array([[float(s) for s in row[0].split()] for i,row in enumerate(reader) if i in lines])
with open(f'{path_pureHE}/001/time.dat', 'r') as fin:
    reader=csv.reader(fin)
    time=np.array([[float(s) for s in row[0].split()] for i,row in enumerate(reader) if i in lines])

for i,line in enumerate(lines):
    fig, axs = plt.subplots(nrows=1,ncols=2, figsize=(20,6))

    plt.suptitle(f'M pure He = {mass[i][0]}, Z=0.01')
    axs[0].plot(time[i],radius[i])
    axs[0].grid(linestyle='dotted',alpha=0.8)
    axs[0].set_xlabel('Time from He ZAMS [Myr]')
    axs[0].set_ylabel('Radius [Rsun]')
    #plt.ylim(0,0.8)
    #plt.show()

    #axs[1].set_title(f'M pure He = {mass[i][0]}, Z=0.02')
    axs[1].plot(time[i],mass[i])
    axs[1].grid(linestyle='dotted',alpha=0.8)
    axs[1].set_xlabel('Time from He ZAMS [Myr]')
    axs[1].set_ylabel('Mass [Msun]')
    #plt.ylim(0,0.8)
    plt.show()
    fig.savefig(f'./parsectracks/pureHE36_Z01_{int(mass[i][0])}.png')
```
